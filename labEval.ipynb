```python
movies=['SpiderMan', 'ToyStory', 'Avengers','Minions', 'BatMan']
```

```python
customer_cnt=0;
customer={}
def createUser():
    name=input("Enter name").lower()
    global customer_cnt
    customer_cnt+=1
    name_id= name+str(customer_cnt)
        
    customer[name_id]=[]
    print("\nUser Created")
    print("Name: ", name)
    print("ID: ", customer_cnt)
    return name_id
    
```

```python
def movieDetails(movie_name):
    name=input("Enter your name: ").lower()
    ids=input("Enter your customer id(if new customer, enter -1)")
    
    movie= movie_name
    name_id=''
    
    if ids=='-1':
        name_id= createUser()
    else:
        
        name_id= name+ids
    customer[name_id].append(movie)
    
    
   # if name_id in customer:
    #    customer[name_id]= customer[name_id].append(movie)
    #else:
     #   customer[name_id]=movie
    
```

```python
def rentMovie():
    print("\nThe Movies Are: \n")
    for i in movies:
        print(i)
        
    movie_name=''
    
    while True:
        movie_name= input("\nEnter the Movie to Rent: ")
        if movie_name in movies:
            print("Redirecting you")
            movieDetails(movie_name)
            movies.remove(movie_name)
            print("\nMovie Rented To You")
            break
        else:
            print("\nPlease choose a Valid Movie")
```

```python
def rentReport():
    name=input("Enter you name:").lower()
    ids=input("enter you id: ")
    name_id=name+ids
    
    if name_id in customer:
        print("\n\n=====Here's Your Current Rented Movies=====")
        print("Name ", name.capitalize(),"\n")
        print("ID ",ids)
     #   print(customer[name_id])
        print("Movies Rented to You:\n")
        for i in customer[name_id]:
            print(i)
    else:
        while True:
            
            print("Customer Not Found\n")
            print("Would you like to create a New Account?\n")
            x= input("y/n:    ")
            if x=='y':
                createUser()
                break
            elif x=='n':
                print("Thankyou\n")
                break
        
            
```

```python
def returnMovie():
    name=input("enter name").lower()
    ids=input("Enter id")
    name_id=name+ids
    
    movie=input("Enter movie to return")
    if name_id in customer:
        try:
            
            customer[name_id].remove(movie)
            movies.append(movie)
            print("Movie Removed Successfully\n")
        except ValueError:
            
   
            print("\nYou do not have this Movie")
```

```python
print("===========Movie Rental System============\n")

print("1.Create Account?\n2.RentMovie  \n3.Return Movie   \n4.Generate Report  \nPress Any Other Character to Exit\n")


while True:
    x=int(input("What do you want to do today?"))
    if x==1:
        createUser()
        continue
    elif x==2:
        rentMovie()
        continue
    elif x==3:
        returnMovie()
        continue
        
    elif x==4:
        rentReport()
        continue
    else:
        break
print("=====Thank You For Your Visit======")
```
